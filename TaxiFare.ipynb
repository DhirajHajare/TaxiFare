```python
df.drop(index_names,inplace=True)
```

```python
df["X1"] = df["latitude_of_pickup"]
df["X2"] = df["latitude_of_dropoff"]
df["Y1"] = df["longitude_of_pickup"]
df["Y2"] = df["longitude_of_dropoff"]
```

```python
df["Distance_Travel"] = np.sqrt((np.square(df["X1"]-df["X2"]))+(np.square(df["Y1"]-df["Y2"])))
```

```python
df1=df.drop(columns=["longitude_of_pickup","latitude_of_pickup","longitude_of_dropoff","latitude_of_dropoff"])
```

```python
df1[["date","Time","Zone"]]=df1["date_time_of_pickup"].str.split(pat=" ",n=-1 , expand=True)
```

```python
df2=df1.copy()
```

```python
df2["amount"] = np.where(df2["amount"]>=52, 52, df2["amount"])
df2["no_of_passenger"] = np.where(df2["no_of_passenger"]>=5, 4, df2["no_of_passenger"])
df2["Distance_Travel"] = np.where(df2["Distance_Travel"]>=0.228218, 0.228218, df2["Distance_Travel"])
```

```python
df2.info()
```

```python
df2["date"]=pd.to_datetime(df2["date"])
df2["date_time_of_pickup"]=pd.to_datetime(df2["date_time_of_pickup"])
```

```python
df2["hours"] = df2["date_time_of_pickup"].dt.hour
df2["min"] = df2["date_time_of_pickup"].dt.minute
df2["second"] = df2["date_time_of_pickup"].dt.second
df2["year"] = df2["date_time_of_pickup"].dt.year
df2["day_of_the_month"] = df2["date_time_of_pickup"].dt.day
```

```python
def time_of_day(x):
    if x in range(6,12):
        return "Morning"
    elif x in range(12,17):
        return "Afternoon"
    elif x in range(17,21):
        return "Evening"
    else:
        return "Night"
```

```python
df2["time_of_day"] = df2["hours"].apply(time_of_day)
df2["day"] = df2["date_time_of_pickup"].dt.day_name()
df2["day_no"] = df2["date_time_of_pickup"].dt.weekday
```

```python
df2 = df2.loc[df2['X1'].between(40, 42)]
df2 = df2.loc[df2['Y1'].between(-75, -72)]
df2 = df2.loc[df2['X2'].between(40, 42)]
df2 = df2.loc[df2['Y2'].between(-75, -72)]
print(f'New number of observations: {df2.shape[0]}')
```

```python
def minkowski_distance(x1, x2, y1, y2, p):
    return ((abs(x2 - x1) ** p) + (abs(y2 - y1)) ** p) ** (1 / p)
```

```python
df2['manhattan_d'] = minkowski_distance(df2['Y1'], 
                                       df2['Y2'],
                                       df2['X1'],
                                       df2['X2'], 1)
```

The Taxifare project is a data science endeavor that focuses on predicting taxi fares. It involves collecting and analyzing a dataset of historical taxi trips, developing a predictive model based on various factors such as distance, time, and location, and evaluating and optimizing the model's performance. The goal is to create an accurate tool for estimating taxi fares, benefiting both taxi service providers and customers.

```python
import pickle
with open(file="random_forest_model.pkl",mode="wb") as f:
    pickle.dump(model5,f)
```

```python
# Assuming you have already trained and saved the Random Forest model as 'random_forest_model.pkl'

# Load the trained model
with open('random_forest_model.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
# Function to preprocess the input data for prediction
def preprocess_input(data):
    processed_data = data  
    
    return processed_data
```

```python
# Function to estimate taxi fares
def estimate_taxi_fare(input_data):
    # Preprocess the input data
    preprocessed_data = preprocess_input(input_data)
    
    # Make predictions using the trained model
    fare_predictions = model.predict(preprocessed_data)
    
    # Return the estimated taxi fares
    return fare_predictions
```

```python
# Estimate taxi fares using the trained model
fare_estimations = estimate_taxi_fare(X_test)

# Display the estimated fares
print(fare_estimations)
```
